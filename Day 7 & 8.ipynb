```python
# Creating a new column or merging 2 columns into one. 
# Example data2['Location'] = data2.City + ', ' + data2.State (ufo.csv in Machine learning Dec)
data_new = pd.read_csv('../Assignments/ufo.csv')
data_new.head(10)
```

```python
data_new['Location'] = data_new['City'] + ', '+ data_new['State']
data_new
```

```python
# To rename columns use the rename() function 
data_new.rename(columns={'Colors Reported': 'Colors_Reported', 'Shape Reported': 'Shape_Reported'}, inplace=True)
# inplace = True will save the changes and make it reflect in the dataset
data_new
```

```python
# Deleting rows: this is done by using the drop() function
# syntax: drop(column_name, axis, inplace). Note that there are two types of axis in dataframe
# the row and the column. the row axis = 0 while the column axis = 1

data_new.drop('Location', axis=1, inplace=True )

data_new 
```

```python
data_new.drop(['City', 'Colors_Reported', 'Shape_Reported'], axis=1, inplace=True)
data_new
```

```python
# Deleting rows:
data_new.drop([3,4], axis=0, inplace=True)
data_new
```
